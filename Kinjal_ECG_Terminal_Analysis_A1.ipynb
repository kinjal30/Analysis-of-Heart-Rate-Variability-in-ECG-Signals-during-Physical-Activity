```python
import matplotlib.pyplot as plt

plt.title("ECG signal - 1000 Hz")
plt.plot(ECG_Data.ECG[0:2000])
plt.xlabel('Time (milliseconds)')
plt.ylabel('Mili Volts')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def annot_max(x, y, ax=None):
    xmax = x[np.argmax(y)]
    ymax = y.max()
    text = "x={:.3f}, y={:.3f}".format(xmax, ymax)
    if not ax:
        ax = plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops = dict(arrowstyle="->", connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data', textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94, 0.96), **kw)

x_values = np.arange(0,1000)
y_values = ECG_Data.ECG.values

plt.title("ECG signal - 1000 Hz")
plt.plot(x_values, y_values[0:1000])
plt.xlabel('Time in Miliseconds')
plt.ylabel('Milli Volts')

annot_max(x_values[0:1000], y_values[0:1000])

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Load the ECG data from a CSV file
data = pd.read_csv('ECG_hw1.csv')
dataRowArray = data['ECG'].to_numpy()

# Assuming a sampling rate of 1000 Hz
Fs = 1000  # Sampling frequency
time = np.arange(len(dataRowArray)) / Fs  # Time in seconds

# R-peak detection
peaks, properties = find_peaks(dataRowArray, height= np.mean(dataRowArray) + np.std(dataRowArray), distance=0.5 * Fs)

# Convert the R-peak indices to time values
time_of_peaks = time[peaks]

# Calculate the heart periods (time between successive R-peaks)
heart_periods = np.diff(time_of_peaks)

# Average Time Calculation for Heart Period Plot
avg_time_between_peaks = (time_of_peaks[:-1] + time_of_peaks[1:]) / 2

# Plot 1: Time vs ECG Amplitude
plt.figure()
plt.plot(time, dataRowArray, label='ECG')
plt.scatter(time[peaks], dataRowArray[peaks], color='r', marker='v', label='R-peaks')
plt.xlabel('Time (s)')
plt.ylabel('ECG Amplitude (mV)')
plt.title('Time vs ECG Amplitude')
plt.legend()
plt.show()

# Plot for the full record: Time vs ECG Amplitude
plt.figure()
plt.plot(time, dataRowArray)
plt.xlabel('Time (s)')
plt.ylabel('ECG Amplitude (mV)')
plt.title('Full Record: Time vs ECG Amplitude')
plt.show()

# Plot for the first 30 seconds: Time vs ECG Amplitude
plt.figure()
plt.plot(time[:30000], dataRowArray[:30000])
plt.xlabel('Time (s)')
plt.ylabel('ECG Amplitude (mV)')
plt.title('First 30 Seconds: Time vs ECG Amplitude')
plt.show()

# Plot 2: Full Record - Time vs Heart Period
plt.figure()
plt.plot(avg_time_between_peaks, heart_periods)
plt.xlabel('Time (s)')
plt.ylabel('Heart Period (s)')
plt.title('Full Record: Time vs Heart Period')
plt.show()

# Plot 3: First 30 Seconds - Time vs Heart Period
idx_30sec_heart_period = avg_time_between_peaks <= 30
plt.figure()
plt.plot(avg_time_between_peaks[idx_30sec_heart_period], heart_periods[idx_30sec_heart_period])
plt.xlabel('Time (s)')
plt.ylabel('Heart Period (s)')
plt.title('First 30 Seconds: Time vs Heart Period')
plt.show()
```

```python
heart_rate_bpm = 60/np.mean(heart_periods)
print(f"Heart Rate: {heart_rate_bpm:.2f} beats per minute")
```
